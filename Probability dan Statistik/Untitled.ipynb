## üß≠ STRUKTUR MODULE: Hypothesis Testing

Berikut urutan materinya (kita nanti bahas satu per satu dengan penjelasan, contoh data science, dan Python):

1. **Introduction to Hypothesis Testing**
   ‚ûú Konsep dasar apa itu hipotesis, null & alternative hypothesis, dan logika pengambilan keputusan.

2. **Inferential Statistics and Hypotheses**
   ‚ûú Bagaimana hipotesis digunakan dalam inferensi; contoh kasus bisnis.

   ‚ûú Cara memilih uji 1 arah atau 2 arah; menghitung test statistic (z atau t).

   ‚ûú Uji proporsi satu populasi (misalnya retention rate atau CTR).

7. **Confidence Interval for the Difference of Means**
   ‚ûú Estimasi selisih dua rata-rata (misalnya rata-rata waktu penggunaan sebelum dan sesudah update).

   ‚ûú Uji hipotesis dua mean (A/B testing, eksperimen produk).

9. **Matched-Pair Hypothesis Testing**
   ‚ûú Uji data berpasangan (misalnya sebelum vs sesudah intervensi untuk user yang sama).

10. **Confidence Interval for the Difference of Proportions**
    ‚ûú Estimasi selisih dua proporsi (misalnya perbandingan CTR dua versi iklan).

11. **Hypothesis Testing for the Difference of Proportions**
    ‚ûú Uji hipotesis perbandingan dua proporsi (proporsi sukses versi A vs B).
