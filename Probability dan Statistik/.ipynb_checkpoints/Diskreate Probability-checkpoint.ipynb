
# Apa itu Discreate Probability

## 1. Probability dasar (Probabilitas biasa)

**Probabilitas dasar** adalah ilmu yang mempelajari seberapa besar kemungkinan suatu **kejadian (event)** terjadi.
Biasanya kita menghitung **peluang dari suatu peristiwa tunggal atau gabungan peristiwa**.

Contohnya:

* Peluang muncul angka 4 saat melempar dadu = 1/6.
* Peluang muncul angka genap (2, 4, 6) = 3/6 = 1/2.
* Peluang muncul dua kepala dalam dua kali lemparan koin = ?

Jadi, **probabilitas dasar** fokus pada **kejadian (event)** yang terjadi dalam satu atau beberapa percobaan.
Belum sampai membahas bagaimana **nilai-nilai acak bisa membentuk distribusi keseluruhan.**

---

## 2. Discrete Probability (Distribusi Probabilitas Diskrit)

Nah, kalau **discrete probability distribution** sudah masuk **tingkat lanjut dari probabilitas dasar.**
Bukan cuma menghitung peluang satu kejadian, tapi **menggambarkan seluruh kemungkinan hasil dari suatu percobaan acak beserta peluangnya masing-masing.**

Coba bedakan:

### Probabilitas dasar:

Kita tanya, ‚ÄúBerapa peluang muncul 3 kepala dari 5 lemparan koin?‚Äù

### Distribusi probabilitas diskrit:

Kita tanya, ‚ÄúBagaimana semua kemungkinan jumlah kepala dari 5 lemparan koin ‚Äî dari 0 sampai 5 ‚Äî beserta peluang masing-masingnya?‚Äù

| Jumlah Kepala (X) | Peluang (P(X)) |
| ----------------- | -------------- |
| 0                 | 1/32           |
| 1                 | 5/32           |
| 2                 | 10/32          |
| 3                 | 10/32          |
| 4                 | 5/32           |
| 5                 | 1/32           |

Nah, **tabel ini** adalah contoh **discrete probability distribution.**

---

## 3. Jadi, perbedaan utamanya:

| Aspek  | Probability Dasar                           | Discrete Probability Distribution                                |
| ------ | ------------------------------------------- | ---------------------------------------------------------------- |
| Fokus  | Menghitung peluang 1 atau beberapa kejadian | Menggambarkan semua kemungkinan hasil dari variabel acak diskrit |
| Hasil  | Nilai tunggal (misal: 0.5, 1/6)             | Sekumpulan nilai peluang untuk setiap nilai X                    |
| Contoh | Peluang muncul angka 4 pada dadu = 1/6      | Peluang setiap sisi dadu (1‚Äì6), semuanya dijumlah jadi 1         |
| Tujuan | Mengetahui peluang dari satu event tertentu | Mengetahui *pola sebaran peluang* semua kemungkinan              |

---

## 4. Kenapa disebut ‚ÄúDiscrete‚Äù?

Karena variabel acaknya hanya bisa memiliki **nilai-nilai yang bisa dihitung (countable)** ‚Äî misalnya 0, 1, 2, 3, ‚Ä¶
Contohnya:

* Jumlah pelanggan yang datang ke toko per jam
* Jumlah error pada sistem per hari
* Jumlah kepala dari lemparan koin

Tidak bisa berupa angka pecahan seperti 1.73 atau 4.12.

Kalau bisa berupa bilangan pecahan, itu masuk **Continuous Probability Distribution** (contohnya distribusi normal).

---

## 5. Analogi sederhana:

Bayangkan kamu melempar dadu 1 kali.

* **Probabilitas dasar:** ‚ÄúBerapa peluang keluar angka 4?‚Äù
  ‚Üí Jawab: 1/6.

* **Distribusi probabilitas diskrit:** ‚ÄúApa peluang masing-masing angka 1, 2, 3, 4, 5, dan 6?‚Äù
  ‚Üí Jawab: semuanya 1/6, totalnya 1.

Distribusi probabilitas memberi **gambaran lengkap dari seluruh peluang yang mungkin**.
Sedangkan probabilitas dasar hanya **mengambil salah satu bagian kecilnya.**

---

## 6. Hubungan dengan data science:

Dalam data science atau machine learning, **discrete probability distribution** sering digunakan untuk:

* Memodelkan jumlah kejadian (seperti klik, error, pesan masuk).
* Memprediksi hasil eksperimen (misalnya jumlah pelanggan yang membeli produk).
* Mengukur kemungkinan hasil kategorikal (misalnya ‚Äúya/tidak‚Äù, ‚Äúlulus/gagal‚Äù).

---

Kalimat simpelnya begini:

> **Probability dasar** = menghitung peluang dari satu kejadian.
> **Discrete probability distribution** = mendaftar semua kejadian yang mungkin *beserta peluangnya.*

### Contoh 5 ‚Äî Kasus Data Science: Jumlah Customer Harian

Misalkan kita memiliki data berikut (jumlah pelanggan harian):

| Hari   | Jumlah Pelanggan (X) |
| ------ | -------------------- |
| Senin  | 3                    |
| Selasa | 4                    |
| Rabu   | 5                    |
| Kamis  | 4                    |
| Jumat  | 6                    |

Kita ingin menghitung **rata-rata, varians, dan standar deviasi**.

```python
import numpy as np

# Data jumlah pelanggan
X = np.array([3, 4, 5, 4, 6])

mean = np.mean(X)
variance = np.var(X, ddof=0)
std_dev = np.std(X, ddof=0)

print(f"Mean (E[X]) = {mean:.2f}")
print(f"Variance = {variance:.2f}")
print(f"Standard Deviation = {std_dev:.2f}")
```

üìà Hasil:

```
Mean (E[X]) = 4.40
Variance = 1.04
Standard Deviation = 1.02
```

**Interpretasi:**
Rata-rata pelanggan per hari adalah 4.4, dengan variasi kecil antar hari (standar deviasi 1.02).

## üîπ 1. Pengertian Discrete Probability

**Probabilitas diskrit** menggambarkan peluang dari variabel acak **diskrit** ‚Äî yaitu variabel yang nilainya **terhitung (countable)** dan terpisah satu sama lain.

Contoh: jumlah pelanggan, jumlah email spam, jumlah heads saat lempar koin, dll.

Syarat dari distribusi probabilitas diskrit:

```
P(X = x_i) ‚â• 0  untuk semua i
Œ£ P(X = x_i) = 1
```

Artinya:

* Semua probabilitas tidak boleh negatif.
* Total semua probabilitas harus sama dengan 1 (karena semua kemungkinan sudah diperhitungkan).

---

## üîπ 2. Probability Mass Function (PMF)

### Apa itu PMF?

**PMF (Probability Mass Function)** adalah fungsi yang menunjukkan peluang bahwa variabel acak diskrit X bernilai tepat sama dengan x tertentu.

```
PMF: f(x) = P(X = x)
```

### Mengapa PMF penting?

Karena PMF memberikan **distribusi peluang individual** ‚Äî berapa besar kemungkinan tiap nilai muncul.
Ini membantu kita memahami **frekuensi kejadian diskrit** dalam data.

---

### Contoh: Lempar Dua Koin ü™ôü™ô

| X (Jumlah Kepala) | P(X=x) |
| ----------------- | ------ |
| 0                 | 0.25   |
| 1                 | 0.50   |
| 2                 | 0.25   |

Jadi fungsi PMF-nya:

```
f(x) =
  0.25 jika x = 0
  0.50 jika x = 1
  0.25 jika x = 2
  0    jika nilai lain
```

Interpretasi:

* Probabilitas mendapatkan 1 kepala = 0.5
* Probabilitas mendapatkan 2 kepala = 0.25

---

## üîπ 3. Cumulative Distribution Function (CDF)

### Apa itu CDF?

**CDF (Cumulative Distribution Function)** adalah fungsi yang menunjukkan **probabilitas kumulatif** sampai nilai tertentu.

```
CDF: F(x) = P(X ‚â§ x) = Œ£ P(X = k) untuk semua k ‚â§ x
```

### Mengapa CDF penting?

CDF menjelaskan **peluang total sampai titik tertentu**.
Cocok digunakan untuk menjawab pertanyaan seperti:

* ‚ÄúBerapa peluang pelanggan membeli ‚â§ 2 kali?‚Äù
* ‚ÄúSeberapa besar kemungkinan error terjadi kurang dari 3 kali?‚Äù

---

### Contoh:

Dari tabel lempar dua koin:

| x | P(X=x) | F(x) = P(X ‚â§ x) |
| - | ------ | --------------- |
| 0 | 0.25   | 0.25            |
| 1 | 0.50   | 0.75            |
| 2 | 0.25   | 1.00            |

Interpretasi:

* P(X ‚â§ 1) = 0.75 ‚Üí peluang mendapatkan maksimal satu kepala adalah 75%.

---

## üîπ 4. Menghitung Peluang Interval

Untuk menghitung peluang nilai dalam rentang tertentu:

```
P(a ‚â§ X ‚â§ b) = P(X=a) + P(X=a+1) + ... + P(X=b)
```

Contoh:

```
P(0 ‚â§ X ‚â§ 1) = P(X=0) + P(X=1) = 0.25 + 0.5 = 0.75
```

---

## üîπ 5. Mengapa PMF dan CDF Penting dalam Data Science

| Aspek                | PMF                                   | CDF                                  |
| -------------------- | ------------------------------------- | ------------------------------------ |
| **Fungsi utama**     | Menunjukkan probabilitas setiap nilai | Menunjukkan probabilitas kumulatif   |
| **Kegunaan**         | Analisis kejadian individu            | Analisis batas probabilitas          |
| **Visualisasi umum** | Grafik batang (bar chart)             | Grafik tangga (step plot)            |
| **Contoh kasus**     | Jumlah klik iklan per hari            | Peluang klik ‚â§ 3 kali                |
| **Insight bisnis**   | Menemukan pola kejadian spesifik      | Menentukan batas ambang probabilitas |

## üîπ 6. Contoh Kasus Data Science üí°

Misal kita punya data e-commerce tentang **jumlah pembelian pelanggan per minggu**.

| Jumlah Pembelian (X) | Peluang (P(X)) |
| -------------------- | -------------- |
| 0                    | 0.2            |
| 1                    | 0.5            |
| 2                    | 0.3            |

Hitung probabilitas pelanggan melakukan pembelian **setidaknya sekali**:

```
P(X ‚â• 1) = P(X=1) + P(X=2)
P(X ‚â• 1) = 0.5 + 0.3 = 0.8
```

Interpretasi:

> Peluang pelanggan berbelanja setidaknya sekali dalam seminggu adalah **80%**.

## üîπ 7. Menghitung dan Visualisasi di Python

```python
import numpy as np
import matplotlib.pyplot as plt

# Nilai dan probabilitas
X = np.array([0, 1, 2])
P_X = np.array([0.2, 0.5, 0.3])

# Hitung CDF
CDF = np.cumsum(P_X)

print("Nilai X:", X)
print("PMF:", P_X)
print("CDF:", CDF)

# Visualisasi
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# PMF
ax[0].bar(X, P_X, color='skyblue')
ax[0].set_title("Probability Mass Function (PMF)")
ax[0].set_xlabel("X")
ax[0].set_ylabel("P(X)")

# CDF
ax[1].step(X, CDF, where='mid', color='orange')
ax[1].set_title("Cumulative Distribution Function (CDF)")
ax[1].set_xlabel("X")
ax[1].set_ylabel("P(X ‚â§ x)")

plt.tight_layout()
plt.show()
```

**Interpretasi hasil Python:**

* Grafik kiri ‚Üí PMF: peluang tiap nilai X.
* Grafik kanan ‚Üí CDF: peluang kumulatif sampai nilai X.

## üîπ 8. Studi Kasus Lengkap: Analisis Transaksi Pelanggan

| Jumlah Transaksi (X) | Frekuensi |
| -------------------- | --------- |
| 0                    | 100       |
| 1                    | 250       |
| 2                    | 150       |
| 3                    | 50        |

Langkah-langkah:

1. Hitung total pelanggan: 100 + 250 + 150 + 50 = 550
2. Hitung PMF dan CDF

```python
import pandas as pd

data = pd.DataFrame({
    'X': [0, 1, 2, 3],
    'freq': [100, 250, 150, 50]
})

data['PMF'] = data['freq'] / data['freq'].sum()
data['CDF'] = data['PMF'].cumsum()

print(data)

plt.figure(figsize=(8,4))
plt.bar(data['X'], data['PMF'], alpha=0.7, label='PMF')
plt.step(data['X'], data['CDF'], where='mid', color='red', label='CDF')
plt.legend()
plt.title('Distribusi Probabilitas Jumlah Pembelian Harian')
plt.xlabel('Jumlah Transaksi (X)')
plt.ylabel('Probabilitas')
plt.show()
```

**Insight:**

* Sebagian besar pelanggan (45%) membeli 1 kali per hari.
* Probabilitas pelanggan membeli ‚â• 2 kali:

  ```
  P(X ‚â• 2) = P(X=2) + P(X=3)
  P(X ‚â• 2) = 0.27 + 0.09 = 0.36
  ```

  ‚ûú Artinya, **36% pelanggan adalah pembeli aktif (lebih dari sekali per hari).**

## üîπ 9. Kesimpulan dan Fungsi PMF & CDF dalam Data Science

| Analisis                 | Fungsi                                                     | Relevansi Data Science                        |
| ------------------------ | ---------------------------------------------------------- | --------------------------------------------- |
| **PMF**                  | Menentukan peluang tiap nilai diskrit                      | Mengetahui distribusi frekuensi kejadian      |
| **CDF**                  | Menentukan peluang kumulatif hingga nilai tertentu         | Menentukan ambang probabilitas (cutoff point) |
| **P(X ‚â• a)**             | Mengetahui peluang kejadian ekstrem atau di atas rata-rata | Mengukur risiko atau peluang tinggi           |
| **Visualisasi (Python)** | Menampilkan pola probabilitas                              | Memudahkan interpretasi & insight bisnis      |



````python
import numpy as np
import matplotlib.pyplot as plt

# Nilai dan probabilitas
X = np.array([0, 1, 2])
P_X = np.array([0.2, 0.5, 0.3])

# Hitung CDF
CDF = np.cumsum(P_X)

print("Nilai X:", X)
print("PMF:", P_X)
print("CDF:", CDF)

# Visualisasi
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# PMF
ax[0].bar(X, P_X, color='skyblue')
ax[0].set_title("Probability Mass Function (PMF)")
ax[0].set_xlabel("X")
ax[0].set_ylabel("P(X)")

# CDF
ax[1].step(X, CDF, where='mid', color='orange')
ax[1].set_title("Cumulative Distribution Function (CDF)")
ax[1].set_xlabel("X")
ax[1].set_ylabel("P(X ‚â§ x)")

plt.tight_layout()
plt.show()
````



# Transforming Random Variables

## 1. Pengantar

**Transforming Random Variables** berarti kita **mengubah variabel acak X menjadi variabel baru Y = g(X)** menggunakan fungsi transformasi tertentu.

Transformasi ini penting karena dalam praktik **data science**, kita sering:

* Mengubah skala data (normalisasi, log transform, square root transform)
* Mengonversi satu distribusi ke distribusi lain (misalnya log-normal)
* Mencari distribusi dari suatu fungsi probabilitas

---

## 2. Konsep Dasar

Jika X adalah **variabel acak diskrit**, dan Y = g(X) adalah fungsi dari X, maka:

* Setiap nilai Y memiliki peluang yang diturunkan dari peluang X.
* Probabilitas untuk Y dihitung dengan menjumlahkan semua probabilitas dari X yang menghasilkan nilai Y tertentu.

**Secara umum:**

P(Y = y) = Œ£ P(X = x) untuk semua x yang menghasilkan g(x) = y

---

## 3. Contoh Matematis

Misalkan X adalah hasil dari lemparan satu dadu (nilai {1, 2, 3, 4, 5, 6}).

Kita buat transformasi:

```
Y = X¬≤
```

Maka nilai Y yang mungkin: {1, 4, 9, 16, 25, 36}

Karena setiap X memiliki probabilitas 1/6, maka:

```
P(Y = 1) = 1/6
P(Y = 4) = 1/6
P(Y = 9) = 1/6
...
P(Y = 36) = 1/6
```

Jadi distribusi probabilitas Y bisa langsung diturunkan dari X.

---

## 4. Tujuan Transformasi

Transformasi dilakukan untuk berbagai alasan:

| Tujuan                     | Contoh Transformasi | Kegunaan                             |
| -------------------------- | ------------------- | ------------------------------------ |
| Mengubah skala             | Y = 10 * X          | Menyesuaikan satuan data             |
| Menstabilkan variansi      | Y = ‚àöX              | Mengurangi pengaruh nilai ekstrem    |
| Mengubah bentuk distribusi | Y = log(X)          | Membuat data lebih mendekati normal  |
| Membuat hubungan linear    | Y = X¬≤              | Membantu analisis regresi non-linear |

---

## 5. Rumus Expected Value Setelah Transformasi

Jika Y = g(X), maka **Expected Value (nilai harapan)** dari Y dihitung dengan:

E(Y) = Œ£ [ g(x) * P(X = x) ]

Artinya, untuk setiap nilai X, kita hitung g(x), kalikan dengan probabilitasnya, lalu jumlahkan semuanya.

---

## 6.Contoh nyata (lengkap): E-commerce ‚Äî dari jumlah pembelian ke revenue per pelanggan

**Konteks (nyata & sederhana):**
Sebuah toko online mengamati jumlah pembelian (X) per pelanggan dalam 1 minggu. Distribusi historisnya adalah:

* P(X = 0) = 0.20
* P(X = 1) = 0.50
* P(X = 2) = 0.30

Pendapatan per pembelian = Rp 50.000. Ada biaya tetap per pelanggan (mis. biaya pengiriman minimum atau biaya layanan) = Rp 10.000.
Jadi definisikan transformasi linear:

```
Y = 50.000 * X + 10.000
```

Y = total revenue (Rp) dari satu pelanggan dalam seminggu.

---

#### 6.1) Hitung PMF Y (perhitungan tangan)

Nilai-nilai X dan P(X):

* X = 0, P = 0.20
* X = 1, P = 0.50
* X = 2, P = 0.30

Gunakan Y = 50.000 * X + 10.000:

* Jika X = 0  ‚Üí Y = 10.000  with P = 0.20
* Jika X = 1  ‚Üí Y = 60.000  with P = 0.50
* Jika X = 2  ‚Üí Y = 110.000 with P = 0.30

Jadi PMF Y:

* P(Y = 10.000)  = 0.20
* P(Y = 60.000)  = 0.50
* P(Y = 110.000) = 0.30

CDF Y (kumulatif):

* F_Y(10.000)  = 0.20
* F_Y(60.000)  = 0.20 + 0.50 = 0.70
* F_Y(110.000) = 1.00

Contoh pertanyaan:

* P(Y ‚â• 60.000) = P(Y = 60.000) + P(Y = 110.000) = 0.50 + 0.30 = 0.80

---

#### 6.2) Hitung E[X], Var(X) dulu (tangan)

E[X] = Œ£ x * P(X = x)

```
E[X] = 0*0.20 + 1*0.50 + 2*0.30 = 0 + 0.50 + 0.60 = 1.10
```

E[X^2] = Œ£ x^2 * P(X = x)

```
E[X^2] = 0^2*0.20 + 1^2*0.50 + 2^2*0.30 = 0 + 0.50 + 1.20 = 1.70
```

Var(X) = E[X^2] - (E[X])^2

```
Var(X) = 1.70 - (1.10)^2 = 1.70 - 1.21 = 0.49
```

Standar deviasi X:

```
SD(X) = sqrt(0.49) = 0.70
```

---

#### 6.3) Gunakan rumus linear untuk E[Y] dan Var(Y]

Karena Y = aX + b dengan a = 50.000, b = 10.000:

```
E[Y] = a * E[X] + b = 50.000 * 1.10 + 10.000 = 55.000 + 10.000 = 65.000  (Rp)
```

```
Var(Y) = a^2 * Var(X) = (50.000)^2 * 0.49
       = 2.500.000.000 * 0.49
       = 1.225.000.000
```

Standar deviasi Y:

```
SD(Y) = sqrt(1.225.000.000) = 35.000  (Rp)
```

**Interpretasi:**

* Rata-rata revenue per pelanggan per minggu = Rp 65.000
* Fluktuasi (SD) sekitar Rp 35.000

---

#### 6.5) Interpretasi Data Science & penggunaan hasil

* Rata-rata revenue per pelanggan (Rp 65k) membantu tim bisnis menentukan **LTV (lifetime value)** kasar per minggu.
* SD besar (Rp 35k) menunjukkan **variabilitas tinggi** antar pelanggan ‚Üí segmen pelanggan perlu dipertimbangkan (mis. banyak pelanggan hanya Rp10k, sebagian besar menghasilkan Rp60k/110k).
* Probabilitas P(Y ‚â• 60.000) = 0.80 ‚Üí 80% pelanggan menghasilkan minimal Rp60k per minggu. Ini bisa jadi input untuk:

  * Capacity planning (estimasi pendapatan harian/mingguan),
  * Simulasi skenario worst/best case untuk cashflow.

#### 6.6) Catatan tambahan (non-linear transformasi singkat)

Jika g bukan linear (mis. Y = X^2 atau Y = log(X+1)), prinsipnya sama:

* Tentukan semua nilai X, hitung Y = g(X) untuk setiap X.
* Kelompokkan nilai Y yang sama (beberapa X bisa map ke nilai Y tunggal) dan jumlahkan probabilitas X yang memetakan ke Y.
* Hitung E[Y] = Œ£ g(x) * P(X = x).
* Var(Y) = Œ£ (g(x))^2 * P(X = x) - (E[Y])^2.

---



````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Distribusi X (jumlah pembelian)
x_values = np.array([0,1,2])
p_x = np.array([0.20, 0.50, 0.30])

# Transformasi Y = a*X + b
a = 50000
b = 10000
y_values = a * x_values + b

# PMF untuk Y (sama mapping karena linear dan one-to-one di sini)
pmf_y = p_x.copy()

# Buat DataFrame ringkas
df = pd.DataFrame({
    'X': x_values,
    'P(X)': p_x,
    'Y (Rp)': y_values,
    'P(Y)': pmf_y
})

# E[X], Var(X) (teoritis)
E_X = np.sum(x_values * p_x)
E_X2 = np.sum((x_values**2) * p_x)
Var_X = E_X2 - E_X**2

# E[Y], Var(Y) using linear property
E_Y = a * E_X + b
Var_Y = (a**2) * Var_X

print("Tabel PMF (X -> Y):")
print(df.to_string(index=False))

print("\nE[X] =", E_X)
print("Var(X) =", Var_X)
print("SD(X) =", np.sqrt(Var_X))

print("\nE[Y] (theoretical) =", E_Y)
print("Var(Y) (theoretical) =", Var_Y)
print("SD(Y) =", np.sqrt(Var_Y))

# Simulasi sampling besar untuk verifikasi
rng = np.random.default_rng(42)
samples_X = rng.choice(x_values, size=200000, p=p_x)
samples_Y = a * samples_X + b

print("\n-- Sample means/vars (from simulation) --")
print("E[X]_sample =", samples_X.mean())
print("Var(X)_sample =", samples_X.var())
print("E[Y]_sample =", samples_Y.mean())
print("Var(Y)_sample =", samples_Y.var())

# Visualisasi PMF dan CDF untuk Y
cdf_y = np.cumsum(pmf_y)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.bar(y_values, pmf_y, width=10000, edgecolor='black')
plt.title('PMF of Y (Revenue per Customer)')
plt.xlabel('Y (Rp)')
plt.ylabel('P(Y)')

plt.subplot(1,2,2)
plt.step(y_values, cdf_y, where='mid', color='red')
plt.scatter(y_values, cdf_y, color='red')
plt.title('CDF of Y')
plt.xlabel('Y (Rp)')
plt.ylabel('P(Y ‚â§ y)')

plt.tight_layout()
plt.show()
````

## 7. Penerapan dalam Data Science

Transformasi variabel sering digunakan dalam:

1. **Feature Engineering**

   * Transformasi log: log(X + 1) untuk mengatasi skewed data
   * Normalisasi: (X - mean) / std untuk membuat data seragam

2. **Stabilizing Variance**

   * Dalam model regresi, kadang hubungan antar variabel tidak linear.
     Solusinya: transformasi X (misalnya log, sqrt) agar hubungan jadi linear.

3. **Handling Outliers**

   * Transformasi akar atau log bisa mengurangi pengaruh nilai ekstrem.

4. **Membentuk Distribusi Baru**

   * Misalnya mengubah data eksponensial menjadi distribusi normal.

---

# Combinations of Random Variables

---

## üîπ 1. Pengantar: Menggabungkan Variabel Acak

Dalam banyak kasus dunia nyata, kita sering tertarik pada **hasil gabungan dari dua variabel acak**.
Contoh:

* Pendapatan total dari **dua cabang toko**: `T = X1 + X2`
* Selisih waktu antara dua kejadian: `D = X1 - X2`
* Rata-rata hasil dua pengukuran sensor: `M = (X1 + X2)/2`

Ketika kita menggabungkan dua variabel acak, kita perlu tahu bagaimana **rata-rata dan varians** berubah.

---

## üîπ 2. Rumus Dasar Kombinasi

Untuk dua variabel acak X dan Y:

```
E[X + Y] = E[X] + E[Y]
E[aX + bY] = aE[X] + bE[Y]
```

Jika **X dan Y independen** (tidak saling memengaruhi):

```
Var(X + Y) = Var(X) + Var(Y)
Var(aX + bY) = a^2 * Var(X) + b^2 * Var(Y)
```

Jika **tidak independen**, tambahkan kovariansi:

```
Var(X + Y) = Var(X) + Var(Y) + 2 * Cov(X, Y)
```

Di mana:

```
Cov(X, Y) = E[(X - E[X]) * (Y - E[Y])]
```

---

## üîπ 3. Contoh 1 ‚Äî Dua Cabang Toko üè¨

Dua toko memiliki distribusi jumlah pembelian harian:

| X (Toko A) | P(X) |
| ---------- | ---- |
| 10         | 0.2  |
| 20         | 0.5  |
| 30         | 0.3  |

| Y (Toko B) | P(Y) |
| ---------- | ---- |
| 5          | 0.3  |
| 15         | 0.4  |
| 25         | 0.3  |

Asumsikan X dan Y **independen**.
Kita ingin mengetahui **total penjualan harian gabungan**:

```
Z = X + Y
```

---

### 3.1 Langkah menghitung PMF Z

Untuk setiap kombinasi (x, y), hitung:

```
P(Z = x + y) = P(X = x) * P(Y = y)
```

Kemudian, **gabungkan nilai Z yang sama** (jumlahkan probabilitasnya).

Misalnya:

* (X=10, Y=5) ‚Üí Z=15, P=0.2*0.3=0.06
* (X=10, Y=15) ‚Üí Z=25, P=0.2*0.4=0.08
* (X=10, Y=25) ‚Üí Z=35, P=0.2*0.3=0.06
* (X=20, Y=5) ‚Üí Z=25, P=0.5*0.3=0.15
* (X=20, Y=15) ‚Üí Z=35, P=0.5*0.4=0.20
* (X=20, Y=25) ‚Üí Z=45, P=0.5*0.3=0.15
* (X=30, Y=5) ‚Üí Z=35, P=0.3*0.3=0.09
* (X=30, Y=15) ‚Üí Z=45, P=0.3*0.4=0.12
* (X=30, Y=25) ‚Üí Z=55, P=0.3*0.3=0.09

Gabungkan Z yang sama:

| Z  | P(Z) |                  |
| -- | ---- | ---------------- |
| 15 | 0.06 |                  |
| 25 | 0.23 | (0.08+0.15)      |
| 35 | 0.35 | (0.06+0.20+0.09) |
| 45 | 0.27 | (0.15+0.12)      |
| 55 | 0.09 |                  |

---

### 3.2 Mean dan Variance

Hitung E[X], E[Y], Var(X), Var(Y) terlebih dahulu:

```
E[X] = 10*0.2 + 20*0.5 + 30*0.3 = 21
E[Y] = 5*0.3 + 15*0.4 + 25*0.3 = 15
Var(X) = (10^2*0.2 + 20^2*0.5 + 30^2*0.3) - (21)^2 = 290 - 441 = 49
Var(Y) = (5^2*0.3 + 15^2*0.4 + 25^2*0.3) - (15)^2 = 175 - 225 = 50
```

Karena independen:

```
E[Z] = E[X] + E[Y] = 36
Var(Z) = Var(X) + Var(Y) = 99
SD(Z) = sqrt(99) ‚âà 9.95
```

Interpretasi:

* Rata-rata total pembelian gabungan = 36 unit per hari
* SD ‚âà 10 unit menunjukkan variasi harian gabungan yang sedang

---

## üîπ 4. Penerapan Data Science

**Konteks:**
Dua toko e-commerce yang menjual produk sama di lokasi berbeda.
Tim data ingin mengetahui total harian gabungan untuk:

* **Forecast penjualan**
* **Inventory planning**
* **Perbandingan antar wilayah**

Menggabungkan dua variabel acak seperti ini adalah dasar untuk:

* **Model agregasi data** (misalnya total penjualan semua cabang)
* **Risk modeling** (total risiko gabungan dua sumber)
* **A/B testing analysis** (membandingkan dua hasil eksperimental independen)
* **Sensor fusion** (menggabungkan dua sensor pengukuran dengan noise berbeda)

## üîπ 5. Implementasi Python

```python
import numpy as np
import pandas as pd
import itertools
import matplotlib.pyplot as plt

# Variabel acak X dan Y
x_vals = np.array([10, 20, 30])
p_x = np.array([0.2, 0.5, 0.3])

y_vals = np.array([5, 15, 25])
p_y = np.array([0.3, 0.4, 0.3])

# Semua kombinasi (x, y)
pairs = list(itertools.product(x_vals, y_vals))

# Hitung nilai Z = X + Y dan probabilitasnya
z_vals = [x + y for x, y in pairs]
p_z = [p_x[list(x_vals).index(x)] * p_y[list(y_vals).index(y)] for x, y in pairs]

# Gabungkan nilai Z yang sama
df = pd.DataFrame({'Z': z_vals, 'P': p_z})
pmf_z = df.groupby('Z')['P'].sum().reset_index()
pmf_z['CDF'] = pmf_z['P'].cumsum()

print("PMF gabungan Z = X + Y")
print(pmf_z)

# Mean dan Variance
E_X = np.sum(x_vals * p_x)
E_Y = np.sum(y_vals * p_y)
Var_X = np.sum((x_vals**2) * p_x) - E_X**2
Var_Y = np.sum((y_vals**2) * p_y) - E_Y**2

E_Z = E_X + E_Y
Var_Z = Var_X + Var_Y

print("\nE[X] =", E_X, "E[Y] =", E_Y)
print("E[Z] =", E_Z)
print("Var(Z) =", Var_Z, "SD(Z) =", np.sqrt(Var_Z))

# Visualisasi PMF dan CDF
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.bar(pmf_z['Z'], pmf_z['P'])
plt.title("PMF of Z = X + Y")
plt.xlabel("Z (Total Pembelian)")
plt.ylabel("P(Z)")

plt.subplot(1,2,2)
plt.step(pmf_z['Z'], pmf_z['CDF'], where='mid', color='red')
plt.title("CDF of Z")
plt.xlabel("Z")
plt.ylabel("P(Z ‚â§ z)")
plt.tight_layout()
plt.show()
```

* Grafik **PMF Z** menunjukkan probabilitas tiap total penjualan (15‚Äì55).
* **CDF Z** naik dari 0 ke 1, memberi probabilitas kumulatif.
* **E[Z] = 36**, artinya rata-rata gabungan dua toko menghasilkan 36 unit per hari.
* Jika target penjualan gabungan harian = 40 unit,
  maka `P(Z ‚â• 40) = 1 - P(Z ‚â§ 35)` dari CDF bisa digunakan untuk **menilai risiko tidak mencapai target.**

2Ô∏è‚É£ Jika Var(X) = 4 dan Var(Y) = 9, dan X, Y independen, berapa Var(Z)?
3Ô∏è‚É£ Mengapa covariance penting jika dua variabel tidak independen?

````python
import numpy as np
import pandas as pd
import itertools
import matplotlib.pyplot as plt

# Variabel acak X dan Y
x_vals = np.array([10, 20, 30])
p_x = np.array([0.2, 0.5, 0.3])

y_vals = np.array([5, 15, 25])
p_y = np.array([0.3, 0.4, 0.3])

# Semua kombinasi (x, y)
pairs = list(itertools.product(x_vals, y_vals))

# Hitung nilai Z = X + Y dan probabilitasnya
z_vals = [x + y for x, y in pairs]
p_z = [p_x[list(x_vals).index(x)] * p_y[list(y_vals).index(y)] for x, y in pairs]

# Gabungkan nilai Z yang sama
df = pd.DataFrame({'Z': z_vals, 'P': p_z})
pmf_z = df.groupby('Z')['P'].sum().reset_index()
pmf_z['CDF'] = pmf_z['P'].cumsum()

print("PMF gabungan Z = X + Y")
print(pmf_z)

# Mean dan Variance
E_X = np.sum(x_vals * p_x)
E_Y = np.sum(y_vals * p_y)
Var_X = np.sum((x_vals**2) * p_x) - E_X**2
Var_Y = np.sum((y_vals**2) * p_y) - E_Y**2

E_Z = E_X + E_Y
Var_Z = Var_X + Var_Y

print("\nE[X] =", E_X, "E[Y] =", E_Y)
print("E[Z] =", E_Z)
print("Var(Z) =", Var_Z, "SD(Z) =", np.sqrt(Var_Z))

# Visualisasi PMF dan CDF
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.bar(pmf_z['Z'], pmf_z['P'])
plt.title("PMF of Z = X + Y")
plt.xlabel("Z (Total Pembelian)")
plt.ylabel("P(Z)")

plt.subplot(1,2,2)
plt.step(pmf_z['Z'], pmf_z['CDF'], where='mid', color='red')
plt.title("CDF of Z")
plt.xlabel("Z")
plt.ylabel("P(Z ‚â§ z)")
plt.tight_layout()
plt.show()
````

## üß† Jawaban kuis Quiz

### **1Ô∏è‚É£ Jika E[X] = 10 dan E[Y] = 20, berapa E[Z] = E[X + Y]?**

**Rumus dasar nilai harapan (Expected Value):**
E[aX + bY] = aE[X] + bE[Y]

Karena di sini a = b = 1 ‚Üí
E[X + Y] = E[X] + E[Y]

**Jawaban:**
E[Z] = 10 + 20 = **30**

**Maknanya:**
Rata-rata dari jumlah dua variabel acak adalah jumlah dari rata-rata masing-masing.
Misalnya:

* X = jumlah pembelian di toko A
* Y = jumlah pembelian di toko B
  ‚Üí Maka rata-rata total pembelian = 30 (gabungan dua toko).

---

### **2Ô∏è‚É£ Jika Var(X) = 4 dan Var(Y) = 9, dan X, Y independen, berapa Var(Z)?**

**Rumus varians untuk penjumlahan dua variabel acak:**

Var(X + Y) = Var(X) + Var(Y) + 2 * Cov(X, Y)

Namun karena **X dan Y independen**, maka **Cov(X, Y) = 0**.

Jadi:
Var(Z) = 4 + 9 = **13**

üß† **Maknanya:**
Varians total adalah penjumlahan dari ketidakpastian masing-masing variabel, selama keduanya tidak saling bergantung.

---

### **3Ô∏è‚É£ Mengapa covariance penting jika dua variabel tidak independen?**

Karena **covariance mengukur seberapa kuat dua variabel bergerak bersama.**

Jika variabel **tidak independen**, maka perubahan pada satu variabel **mempengaruhi** yang lain.
Covariance memberi tahu:

* Apakah hubungan itu **positif** (bergerak searah)
* **Negatif** (bergerak berlawanan)
* Atau **tidak ada hubungan linear** (mendekati nol)

üß© **Contoh Data Science:**
Dalam **analisis portofolio saham**, jika dua saham punya covariance positif tinggi, risikonya meningkat karena keduanya cenderung naik/turun bersamaan.
Tapi jika covariance-nya negatif, kamu bisa **diversifikasi risiko**.

---

### **4Ô∏è‚É£ Bagaimana kamu akan menerapkan kombinasi variabel acak untuk memprediksi total demand dari dua cabang toko berbeda?**

#### **Konsepnya:**

Misal:

* X = permintaan (demand) di toko cabang A
* Y = permintaan (demand) di toko cabang B
  ‚Üí Maka total demand Z = X + Y

Untuk memprediksi total demand, kita gunakan:

* **E[Z] = E[X] + E[Y]** ‚Üí untuk rata-rata total permintaan
* **Var(Z) = Var(X) + Var(Y) + 2Cov(X,Y)** ‚Üí untuk mengukur seberapa tidak pastinya total demand

Toko A dan B menjual produk yang sama.

| Cabang | Mean Demand | Variance | Covariance |
| ------ | ----------- | -------- | ---------- |
| A      | 100         | 25       | ‚Äî          |
| B      | 120         | 36       | 20         |

Maka:

* E[Z] = 100 + 120 = **220**
* Var(Z) = 25 + 36 + 2(20) = **101**

Artinya:

* Rata-rata total permintaan: 220 unit
* Variansi total permintaan: 101 unit¬≤ (menggambarkan ketidakpastian)


**Insight:**

* Kamu bisa melihat **berapa rata-rata permintaan gabungan**.

# Permutasi dan Kombinasi
**Permutations and Combinations**, yaitu konsep yang sangat penting dalam **probabilitas diskrit dan data science**, terutama saat menghitung **jumlah kemungkinan (possible outcomes)** dari suatu peristiwa acak.



##  **1. Pengantar: Kenapa Permutations & Combinations Penting?**

Dalam **probabilitas diskrit**, kita sering ingin tahu:

> ‚ÄúDari sekian banyak kemungkinan hasil, berapa banyak cara sesuatu bisa terjadi?‚Äù

Misalnya:

* Berapa banyak cara memilih **3 produk unggulan** dari **10 produk**?
* Berapa banyak cara menyusun **password 4 huruf** dari alfabet?
* Dalam dataset, berapa banyak **kombinasi fitur** yang bisa dipilih untuk model machine learning?

Nah, semua pertanyaan itu diselesaikan dengan **permutasi dan kombinasi**.

---

##  **2. Permutations (Permutasi)**

###  **Definisi:**

Permutasi adalah **cara menyusun** objek di mana **urutan (order) penting**.

> Rumus:
> P(n, r) = n! / (n - r)!

Keterangan:

* **n** = jumlah total objek
* **r** = jumlah objek yang diambil
* **!** (faktorial) berarti perkalian berurutan dari n ke 1 (contoh: 5! = 5√ó4√ó3√ó2√ó1)

---

###  **Contoh 1:**

Berapa banyak cara menyusun 3 huruf pertama dari kata **DATA**?

* Total huruf (n) = 4
* Dipilih (r) = 3
* Maka:
  P(4,3) = 4! / (4 - 3)! = 24 / 1 = **24 cara**

---

###  **Contoh Data Science:**

Jika kamu memiliki **5 model machine learning**, dan ingin menyusun **3 model terbaik untuk voting ensemble**,
urutan penyusunan model **berpengaruh pada hasil akhir** (karena urutan menentukan bobot voting).

Jadi gunakan **permutasi**:
P(5,3) = 5! / (2)! = 60 cara.

---

###  **Python Implementation**

```python
import math

# Permutation P(n, r)
n = 5
r = 3
permutasi = math.perm(n, r)
print("Jumlah permutasi:", permutasi)
```

Output:

```
Jumlah permutasi: 60
```

---

##  **3. Combinations (Kombinasi)**

###  **Definisi:**

Kombinasi adalah **cara memilih objek** di mana **urutan tidak penting**.

> Rumus:
> C(n, r) = n! / [r! √ó (n - r)!]

Keterangan:

* **Urutan tidak diperhatikan**.
* Biasanya digunakan saat ‚Äúmemilih‚Äù, bukan ‚Äúmenyusun‚Äù.

---

###  **Contoh 2:**

Berapa banyak cara memilih **3 mahasiswa** dari **10** untuk proyek?

C(10,3) = 10! / (3! √ó 7!) = 120

Artinya, ada **120 kombinasi** berbeda dari 3 mahasiswa yang bisa dipilih tanpa memperhatikan urutan.

---

###  **Contoh Data Science:**

Dalam pemilihan **fitur model (feature selection)**, kita tidak peduli urutan fitur:

* Dari 6 fitur, berapa banyak kombinasi 3 fitur bisa diuji?
  C(6,3) = 6! / (3! √ó 3!) = 20 kombinasi.

Model machine learning bisa dilatih dengan **20 subset fitur berbeda**.

---

###  **Python Implementation**

```python
import math

# Combination C(n, r)
n = 6
r = 3
kombinasi = math.comb(n, r)
print("Jumlah kombinasi:", kombinasi)
```

Output:

```
Jumlah kombinasi: 20
```

---

##  **4. Hubungan dengan Probabilitas**

Permutasi & kombinasi digunakan untuk menghitung **probabilitas diskrit**.

Misalnya, probabilitas memilih 2 kartu As dari 5 kartu:

* Total kombinasi 2 kartu dari 5: C(5,2) = 10
* Kombinasi kartu As yang mungkin: C(2,2) = 1
  ‚Üí Probabilitas = 1 / 10 = 0.1 (10%)

# **Binomial Random Variables dan Binomial Distribution**

---

## 1. Pengantar

Dalam statistik dan data science, kita sering menghadapi situasi dengan **dua kemungkinan hasil** ‚Äî seperti *sukses/gagal*, *beli/tidak beli*, *klik/tidak klik*, *lulus/tidak lulus*, dll.

Nah, model dasar dari situasi seperti ini disebut **Bernoulli trial**.

Jika percobaan ini dilakukan **berulang kali sebanyak n kali** dan kita ingin tahu **berapa banyak sukses yang terjadi**,
maka jumlah keberhasilan itu disebut:

```
Binomial Random Variable
```

---

## 2. Definisi Formal

Sebuah variabel acak **X** dikatakan **berdistribusi binomial** jika:

* Percobaan dilakukan **n kali** secara independen.
* Setiap percobaan punya dua hasil: sukses (probabilitas = p) atau gagal (probabilitas = 1 - p).
* **X** menghitung **jumlah keberhasilan** dalam n percobaan tersebut.

Ditulis sebagai:

```
X ~ Binomial(n, p)
```

Artinya:

> Variabel acak X mengikuti **Binomial Distribution** dengan parameter n dan p.

---

## 3. Hubungan Antara Binomial Random Variable dan Binomial Distribution

| Konsep                           | Pengertian                                                              | Hubungan                    |
| -------------------------------- | ----------------------------------------------------------------------- | --------------------------- |
| **Binomial Random Variable (X)** | Variabel acak yang menghitung jumlah sukses dari n percobaan Bernoulli. | Objek yang kita ukur.       |
| **Binomial Distribution**        | Distribusi peluang dari semua kemungkinan nilai X (0, 1, 2, ..., n).    | Sifat probabilistik dari X. |

**Kesimpulan:**

> ‚ÄúBinomial Random Variable adalah apa yang diukur, sedangkan Binomial Distribution adalah bagaimana nilai-nilai variabel itu tersebar secara probabilistik.‚Äù

---

## 4. Rumus Probability Mass Function (PMF)

Distribusi binomial punya fungsi peluang (PMF) seperti berikut:

```
P(X = k) = C(n, k) * p^k * (1 - p)^(n - k)
```

di mana:

* `C(n, k)` = kombinasi n pilih k = n! / (k! * (n - k)!)
* `p` = peluang sukses
* `1 - p` = peluang gagal
* `k` = jumlah sukses yang ingin dihitung (0 ‚â§ k ‚â§ n)

---

## 5. Mean, Variance, dan Standard Deviation

Rata-rata dan penyebaran Binomial Distribution bisa dihitung dengan:

```
Mean (E[X]) = n * p
Variance (Var[X]) = n * p * (1 - p)
Standard Deviation (SD[X]) = ‚àö(n * p * (1 - p))
```

Interpretasi:

* **E[X]** = rata-rata keberhasilan yang diharapkan.
* **Var[X]** = seberapa besar variasinya.
* **SD[X]** = seberapa jauh nilai cenderung menyimpang dari rata-rata.

---

## 6. Contoh Nyata di Dunia Data Science

### Kasus:

Sebuah perusahaan e-commerce memiliki peluang **p = 0.2** bahwa seorang pengunjung akan membeli produk.
Dalam satu jam, ada **n = 10 pengunjung**.
Kita ingin tahu:

1Ô∏è‚É£ Peluang bahwa **tepat 3 orang membeli**
2Ô∏è‚É£ Berapa rata-rata (mean) dan deviasi standarnya

---

### Langkah Perhitungan Manual

```
P(X = 3) = C(10, 3) * (0.2)^3 * (0.8)^7
          = 120 * 0.008 * 0.2097
          = 0.2013
```

üìç Jadi peluang **tepat 3 pembelian dari 10 pengunjung** = **20.13%**.

### üîπ Hitung Mean dan Variance

```
E[X] = n * p = 10 * 0.2 = 2
Var[X] = n * p * (1 - p) = 10 * 0.2 * 0.8 = 1.6
SD[X] = ‚àö1.6 = 1.26
```

Interpretasi:

> Rata-rata diharapkan 2 pembelian, dengan penyebaran sekitar 1.26.

---

## 7. Implementasi Python

```python
from math import comb

```


## **Poisson Distribution dan Poisson Random Variable**

###  **Pengantar**

Distribusi **Poisson** digunakan untuk memodelkan **jumlah kejadian (events)** yang terjadi dalam suatu **interval waktu, ruang, atau area tertentu**, dengan syarat:

* Kejadian **terjadi secara independen**.
* Kejadian **terjadi dengan rata-rata tertentu (Œª)** per interval.
* Peluang dua kejadian terjadi pada waktu yang **sangat berdekatan ‚âà 0**.

Contoh nyata:

* Jumlah email masuk per jam.
* Jumlah mobil melewati tol per menit.
* Jumlah error pada server per hari.
* Jumlah pelanggan datang ke toko per jam.

---

###  **Rumus Poisson Distribution**

P(X = k) = (e^(-Œª) * Œª^k) / k!

Keterangan:

* **X** = jumlah kejadian dalam interval tertentu
* **Œª (lambda)** = rata-rata kejadian per interval
* **k** = jumlah kejadian yang diinginkan (0, 1, 2, ‚Ä¶)
* **e** = konstanta Euler ‚âà 2.71828

---

###  **Sifat-Sifat Penting**

* **Mean (E[X]) = Œª**

* **Variance (Var[X]) = Œª**
  Jadi mean dan variance sama!

* Cocok untuk proses diskrit dengan skala waktu atau ruang kontinu.

* Poisson sering menjadi **aproksimasi dari distribusi binomial** jika:

  * n besar (n > 20)
  * p kecil (p < 0.05)
  * Œª = n * p

---

###  **Contoh Nyata: Jumlah Email per Jam**

Misalkan rata-rata **Œª = 3 email/jam**.

#### (a) Berapa peluang menerima tepat 5 email dalam 1 jam?

Gunakan rumus:
P(X = 5) = (e^(-3) * 3^5) / 5!
= (0.0498 * 243) / 120
‚âà 0.1008

‚û°Ô∏è **P(X = 5) = 0.1008 (10.08%)**

---

#### (b) Peluang menerima paling banyak 2 email:

P(X ‚â§ 2) = P(0) + P(1) + P(2)

* P(0) = e^(-3) * 3^0 / 0! = 0.0498
* P(1) = e^(-3) * 3^1 / 1! = 0.1494
* P(2) = e^(-3) * 3^2 / 2! = 0.2240

P(X ‚â§ 2) = 0.0498 + 0.1494 + 0.2240 = **0.4232**

‚û°Ô∏è **P(X ‚â§ 2) = 42.32%**

---

### **Hubungan Poisson dan Binomial**

Distribusi **Poisson** adalah **limit dari distribusi binomial** ketika:

* n ‚Üí ‚àû
* p ‚Üí 0
* Œª = n * p (konstan)

Jadi Poisson cocok untuk kasus ketika kita tidak tahu jumlah percobaan pasti, tapi tahu **rata-rata kejadian**.

---

# "At least" dan "At most", serta Mean, Variance, dan Standard Deviation

## 1) Inti konsep: "at most" vs "at least"


| Istilah      | Arti harfiah                       | Arti probabilitas                                                   |
| ------------ | ---------------------------------- | ------------------------------------------------------------------- |
| **At most**  | ‚Äúpaling banyak‚Äù                    | Probabilitas bahwa nilai X tidak melebihi angka tertentu (X ‚â§ k)    |
| **At least** | ‚Äúsetidaknya‚Äù atau ‚Äúpaling sedikit‚Äù | Probabilitas bahwa nilai X tidak kurang dari angka tertentu (X ‚â• k) |

---

* **At most k**
  Artinya semua kemungkinan **sampai** nilai k dihitung:

  ```
  P(X ‚â§ k) = P(X=0) + P(X=1) + ... + P(X=k)
  ```

* **At least k**
  Artinya semua kemungkinan **dari k ke atas** dihitung:

  ```
  P(X ‚â• k) = P(X=k) + P(X=k+1) + ... + P(X=n)
  ```

  Tapi biasanya lebih efisien dihitung dengan komplemen:

  ```
  P(X ‚â• k) = 1 ‚àí P(X ‚â§ k‚àí1)
  ```


Trik praktis: gunakan CDF untuk menghitung "at most" langsung; gunakan komplemen CDF untuk "at least".

---

## 2) Rumus umum (variabel acak diskrit)

* PMF: `P(X = x) = f(x)` (diberi/terhitung)
* CDF: `F(k) = P(X ‚â§ k) = Œ£ f(x)` untuk semua x ‚â§ k
* "At most": `P(X ‚â§ k) = F(k)`
* "At least": `P(X ‚â• k) = 1 - F(k - 1)`

Contoh alternatif (saat butuh P(a ‚â§ X ‚â§ b)):

```
P(a ‚â§ X ‚â§ b) = F(b) - F(a - 1)
```

---

## 3) Mean, Variance, Standard Deviation ‚Äî definisi & rumus praktis

**General discrete case**

* Expected value (mean):

  ```
  E[X] = Œ£ x * P(X = x)  over all possible x
  ```
* Second moment:

  ```
  E[X^2] = Œ£ x^2 * P(X = x)
  ```
* Variance:

  ```
  Var(X) = E[X^2] - (E[X])^2
  ```
* Standard deviation:

  ```
  SD(X) = sqrt(Var(X))
  ```

**Formula khusus (cepat)**

* Binomial(n, p):

  ```
  Mean = n * p
  Variance = n * p * (1 - p)
  SD = sqrt(n * p * (1 - p))
  ```
* Poisson(lambda):

  ```
  Mean = lambda
  Variance = lambda
  SD = sqrt(lambda)
  ```

---

## 4) Contoh lengkap ‚Äî Binomial (praktik "at least" / "at most")

**Kasus**: n = 10, p = 0.3.

1. Hitung `P(X ‚â§ 2)` (at most 2): gunakan CDF langsung.
2. Hitung `P(X ‚â• 3)` (at least 3): `1 - P(X ‚â§ 2)`.

**Perhitungan Python (scipy)**

```python
from scipy.stats import binom

n = 10
p = 0.3

# at most 2
p_at_most_2 = binom.cdf(2, n, p)

# at least 3
p_at_least_3 = 1 - binom.cdf(2, n, p)  # = 1 - P(X <= 2)

print("P(X <= 2) =", p_at_most_2)
print("P(X >= 3) =", p_at_least_3)

# mean and variance
mean = n * p
var = n * p * (1 - p)
sd = var ** 0.5
print("Mean:", mean, "Variance:", var, "SD:", sd)
```

**Interpretasi**:

* `P(X ‚â§ 2)` menunjukkan probabilitas ‚Äúpendek‚Äù (few successes).
* `P(X ‚â• 3)` adalah peluang bahwa jumlah sukses cukup banyak (3 atau lebih).
* Mean = 3, Var = 2.1, SD ‚âà 1.449.

---

## 5) Contoh lengkap ‚Äî Poisson (praktik "at most" / "at least")

**Kasus**: lambda = 3 (rata-rata 3 kejadian per interval).

1. `P(X ‚â§ 2)` = probabilitas paling banyak 2 kejadian.
2. `P(X ‚â• 5)` = probabilitas setidaknya 5 kejadian = `1 - P(X ‚â§ 4)`.

**Perhitungan Python (scipy)**

```python
from scipy.stats import poisson

lmbda = 3

p_at_most_2 = poisson.cdf(2, lmbda)
p_at_least_5 = 1 - poisson.cdf(4, lmbda)

print("Poisson Œª=3: P(X <= 2) =", p_at_most_2)
print("Poisson Œª=3: P(X >= 5) =", p_at_least_5)

# mean & var
print("Mean = Variance = ", lmbda, "SD =", lmbda**0.5)
```

**Interpretasi**:

* Gunakan `cdf` untuk "‚â§" dan komplemen untuk "‚â•".
* Untuk Poisson, mean dan variance sama (lambda), jadi SD = sqrt(lambda).

---

## 6) Calculate mean/var from PMF (general example)

Misal PMF diberikan tabel:

```
x: 0   1    2
P: 0.2 0.5  0.3
```

Python:

```python
import numpy as np

x = np.array([0,1,2])
p = np.array([0.2,0.5,0.3])

E = np.sum(x * p)
E2 = np.sum((x**2) * p)
Var = E2 - E**2
SD = np.sqrt(Var)

print("E =", E)
print("Var =", Var)
print("SD =", SD)
```

---

## 7) Kenapa ini penting di Data Science? (interpretasi praktis)

* **"At most" / "At least"** sering muncul di bisnis sebagai ambang keputusan:

  * "Apa peluang penjualan ‚â§ target?" (risiko underperformance)
  * "Apa peluang pelanggan ‚â• threshold?" (peluang high-value customers)
* **Mean** digunakan untuk estimasi rata-rata (forecast).
* **Variance/SD** mengukur ketidakpastian/risiko: planning inventory, buffer, kapasitas.
* Kombinasi contoh: gunakan `P(total demand ‚â• capacity)` untuk capacity planning; gunakan SD untuk membuat safety stock.

---

## 8) Contoh kasus terpadu (Data Science)

**Kasus**: E-commerce, p = 0.15 konversi per pengunjung, n = 200 kunjungan per hari.

1. Mean konversi per hari = n*p = 30.
2. SD = sqrt(n*p*(1-p)) ‚âà sqrt(200*0.15*0.85) ‚âà 5.03.
3. Peluang at least 40 pembelian = 1 ‚àí P(X ‚â§ 39) (pakai binom.cdf).

Python:

```python

n, p = 200, 0.15
mean = n*p
sd = (n*p*(1-p))**0.5
p_at_least_40 = 1 - binom.cdf(39, n, p)

print("Mean:", mean, "SD:", sd)
print("P(X >= 40) =", p_at_least_40)
```

Interpretasi: jika P(X ‚â• 40) kecil, target 40 terlalu optimis; jika besar, target realistis.

---

## 9) Ringkasan langkah cepat untuk soal "at least / at most"

* If you need P(X ‚â§ k): compute CDF at k.
* If you need P(X ‚â• k): compute 1 ‚àí CDF(k ‚àí 1).
* If you need P(a ‚â§ X ‚â§ b): compute CDF(b) ‚àí CDF(a ‚àí 1).
* For mean/var/SD, gunakan rumus khusus (Binomial/Poisson) atau hitung dari PMF secara umum.

# Bernoulli Random Variables

---

## Apa itu Bernoulli Random Variable?

**Bernoulli random variable** adalah **variabel acak yang hanya punya dua kemungkinan hasil: sukses atau gagal.**

Contohnya:

* Melempar koin ‚Üí kepala (1) atau ekor (0)
* Pelanggan membeli produk ‚Üí ya (1) atau tidak (0)
* Email diklik ‚Üí ya (1) atau tidak (0)

Jadi, nilai X hanya bisa:

```
X = 1 (sukses)
X = 0 (gagal)
```

---

## Probability Function (PMF)

Distribusi probabilitas Bernoulli ditulis sebagai:

```
P(X = 1) = p
P(X = 0) = 1 - p
```

Secara umum:

```
P(X = x) = p^x * (1 - p)^(1 - x)  untuk x = 0, 1
```

---

## Parameter

* **p** = probabilitas sukses (0 ‚â§ p ‚â§ 1)
* **1 - p** = probabilitas gagal

---

## Mean, Variance, dan Standard Deviation

Karena hanya dua hasil, perhitungannya sederhana:

```
Mean (E[X]) = p
Variance (Var[X]) = p * (1 - p)
Standard Deviation (SD[X]) = sqrt(p * (1 - p))
```

Maknanya:

* **Mean** adalah peluang sukses (nilai harapan).
* **Variance/SD** menunjukkan seberapa tidak pasti hasilnya (jika p mendekati 0.5, variasi lebih tinggi).

---

## Contoh 1 ‚Äî Melempar koin

Misal:

* X = 1 jika muncul kepala.
* p = 0.5 (peluang sukses)

Maka:

```
E[X] = 0.5
Var[X] = 0.5 * (1 - 0.5) = 0.25
SD[X] = sqrt(0.25) = 0.5
```

Interpretasi:

* Rata-rata (nilai harapan) dari satu kali lemparan = 0.5 (50% sukses)
* SD 0.5 menunjukkan hasilnya cukup bervariasi (kadang 0, kadang 1).

---

## Contoh 2 ‚Äî Data Science Case: Conversion Rate

**Kasus:**
Sebuah iklan online memiliki peluang 0.2 bahwa seorang pengunjung akan mengklik (click).

Artinya:

```
p = 0.2, X = 1 jika klik, X = 0 jika tidak.
```

Perhitungan:

```
E[X] = 0.2
Var[X] = 0.2 * 0.8 = 0.16
SD[X] = sqrt(0.16) = 0.4
```

Interpretasi:

* Rata-rata 20% pengunjung akan klik.
* Ada variabilitas cukup besar antar individu (SD 0.4).

---


## Simulasi dan Estimasi di Data Science

## Hubungan Bernoulli dan Binomial

| Distribusi    | Karakteristik                                       | Hubungan                  |
| ------------- | --------------------------------------------------- | ------------------------- |
| **Bernoulli** | 1 percobaan (n=1), 2 hasil                          | Dasar dari binomial       |
| **Binomial**  | n percobaan independen, tiap percobaan Bernoulli(p) | Jumlahkan hasil Bernoulli |

**Binomial = jumlah dari banyak Bernoulli.**

Artinya:

```
Jika X‚ÇÅ, X‚ÇÇ, ‚Ä¶, X‚Çô ~ Bernoulli(p), maka Œ£Xi ~ Binomial(n, p)
```

# Geometric Random Variables

---

##  Apa itu Geometric Random Variable?

**Geometric random variable** adalah variabel acak yang menghitung **berapa kali percobaan Bernoulli harus dilakukan sampai mendapatkan sukses pertama.**

üëâ Jadi, kalau Bernoulli menghitung ‚Äúapakah sukses atau gagal‚Äù,
maka **Geometric menghitung berapa banyak percobaan yang dibutuhkan untuk mendapatkan sukses pertama.**

---

### Contoh sederhana

Misal kita melempar koin:

* X = jumlah lemparan sampai kepala pertama muncul.
* p = peluang muncul kepala = 0.5.

Kemungkinan hasil:

```
X = 1 ‚Üí sukses di lemparan pertama.
X = 2 ‚Üí gagal dulu (ekor), lalu sukses (kepala).
X = 3 ‚Üí dua kali gagal, lalu sukses.
dst...
```

---

## Probability Mass Function (PMF)

Fungsi probabilitas dari X adalah:

```
P(X = k) = (1 - p)^(k - 1) * p   untuk k = 1, 2, 3, ...
```

Artinya:

* Gagal (1 - p) sebanyak (k - 1) kali,
* lalu sukses (p) di percobaan ke-k.

---

## Mean, Variance, dan Standard Deviation

Rumus dasarnya:

```
E[X] = 1 / p
Var[X] = (1 - p) / p^2
SD[X] = sqrt((1 - p) / p^2)
```

Interpretasi:

* Semakin kecil p, semakin lama rata-rata waktu menunggu sukses pertama.
* Variansi besar kalau peluang sukses kecil ‚Üí makin tidak pasti.

---

## Contoh 1 ‚Äî Lempar koin

Misal p = 0.5 (peluang kepala).
Berapa ekspektasi jumlah lemparan untuk dapat kepala pertama?

```
E[X] = 1 / 0.5 = 2
Var[X] = (1 - 0.5) / (0.5)^2 = 0.5 / 0.25 = 2
SD[X] = sqrt(2) = 1.414
```

Artinya rata-rata butuh 2 lemparan untuk mendapatkan kepala pertama.

## Contoh 2 ‚Äî Data Science Case: Conversion Rate

üìä **Kasus:**
Kita ingin tahu rata-rata berapa kali iklan perlu ditampilkan sampai ada 1 klik.
Jika peluang klik per iklan = 0.1, maka:

```
p = 0.1
E[X] = 1 / 0.1 = 10 ‚Üí rata-rata 10 tayangan untuk dapat 1 klik
Var[X] = (1 - 0.1) / (0.1)^2 = 0.9 / 0.01 = 90
SD[X] = sqrt(90) = 9.49
```

Jadi secara rata-rata dibutuhkan **10 tampilan** untuk satu klik,
tapi hasilnya bervariasi cukup tinggi (SD = 9.49).

## Visualisasi PMF dengan Python

```python
import matplotlib.pyplot as plt
from scipy.stats import geom
import numpy as np

p = 0.2
x = np.arange(1, 15)
pmf = geom.pmf(x, p)

plt.stem(x, pmf, use_line_collection=True)
plt.title(f"Geometric Distribution (p = {p})")
plt.xlabel("X = jumlah percobaan sampai sukses pertama")
plt.ylabel("P(X=x)")
plt.show()
```

Grafiknya akan menunjukkan probabilitas menurun tajam ‚Äî
karena semakin lama suksesnya, semakin kecil kemungkinannya.

## Simulasi dalam Python

Simulasikan 10000 percobaan untuk melihat hasil rata-rata empiris:

```python
import numpy as np

p = 0.2
n = 10000
samples = np.random.geometric(p, size=n)

print("Mean empiris:", np.mean(samples))
print("Mean teoritis:", 1/p)
```

üìà Hasil mean empiris akan mendekati nilai teoritis (‚âà 5 jika p=0.2).

**Geometric** = deret dari percobaan Bernoulli sampai sukses pertama.

## Kesimpulan

| Konsep   | Rumus                     | Arti                                     |
| -------- | ------------------------- | ---------------------------------------- |
| PMF      | P(X=k) = (1-p)^(k-1)p     | Peluang sukses di percobaan ke-k         |
| Mean     | E[X] = 1/p                | Rata-rata jumlah percobaan sampai sukses |
| Variance | Var[X] = (1-p)/p¬≤         | Ukur ketidakpastian waktu sukses         |
| SD       | sqrt((1-p)/p¬≤)            | Simpangan baku                           |
| Aplikasi | Menunggu kejadian pertama | Konversi, kegagalan, pembelian, dll      |
