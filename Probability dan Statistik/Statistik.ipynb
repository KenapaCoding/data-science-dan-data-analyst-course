# ğŸ“Š Measures of Center (Ukuran Pemusatan)

## ğŸ”¹ Pengertian
Ukuran pemusatan digunakan untuk mengetahui **nilai tengah atau pusat** dari sekelompok data.  
Tiga ukuran utama adalah:
1. **Mean (Rata-rata)**
2. **Median**
3. **Mode (Modus)**

## Mean

## Apa itu Mean?
**Mean (Rata-rata)** adalah salah satu ukuran pemusatan data yang paling sering digunakan.  
Mean dihitung dengan menjumlahkan semua nilai data, kemudian dibagi dengan banyaknya data.  

### Rumus Mean

Mean = (Î£ xáµ¢) / n  
di mana:
- Î£ xáµ¢ = jumlah semua nilai data
- n = banyaknya data

---

## âœ¨ Contoh Perhitungan (langkah demi langkah)
Misalkan data nilai: `70, 80, 90, 100`

1. Jumlahkan semua nilai:  
   `Î£ xáµ¢ = 70 + 80 + 90 + 100 = 340`

2. Hitung banyak data:  
   `n = 4`

3. Hitung mean: Î£ xáµ¢ / n= 340 / 4

Jadi, rata-rata nilai ujian = **85**.

---

# Measure of Center: Median

## ğŸ“Œ Apa itu Median?
**Median** adalah nilai tengah dari suatu distribusi data yang telah diurutkan.  
- 50% data berada **di bawah median**  
- 50% data berada **di atas median**  
- Median juga disebut **50th percentile** atau titik tengah distribusi.

---

## ğŸ§® Cara Menentukan Median
1. Urutkan data dari kecil ke besar.  
2. Jika jumlah data **ganjil**, median = nilai di tengah.  
3. Jika jumlah data **genap**, median = rata-rata dari dua nilai tengah.

---

## âœ¨ Contoh Perhitungan Manual

### 1) Jumlah Data Ganjil
Data usia: `15, 23, 18, 15, 24, 23, 17`

1. Urutkan data â†’ `15, 15, 17, 18, 23, 23, 24`  
2. Jumlah data \(n = 7\) (ganjil).  
3. Nilai tengah ke-4 adalah **18**.  

â¡ï¸ **Median = 18 tahun**

---

### 2) Jumlah Data Genap
Data harga rumah (ribu USD): `325, 375, 385, 395, 420, 825`

1. Data sudah urut.  
2. Jumlah data \(n = 6\) (genap).  
3. Dua nilai tengah = `385` dan `395`.  
4. Median = (385 + 395) / 2 = **390**.  

â¡ï¸ **Median = $390,000**

---

## ğŸ“Š Perbandingan dengan Mean
- **Mean (rata-rata)** bisa terpengaruh oleh **outlier** (nilai ekstrem).  
- **Median** lebih tahan terhadap outlier.  

Contoh:  
- Data harga rumah di atas punya **mean = 454,167** (lebih besar karena ada outlier 825).  
- Tapi **median = 390** yang lebih merepresentasikan harga tipikal.

---

# Measures of Center: Outliers & Distribution Shapes

## ğŸ“Œ Outliers
### Apa itu Outlier?
**Outlier** adalah nilai data yang sangat berbeda dari data lainnya, jauh dari pusat distribusi.  
- Outlier dapat menarik **Mean** ke arah nilai ekstrem.  
- **Median** dan **Mode** lebih tahan terhadap outlier (resistant measures).  
- Oleh karena itu, jika data memiliki nilai ekstrem, **Median** lebih baik digunakan daripada Mean.

### Contoh Kasus
Sebuah perusahaan ingin melaporkan **rata-rata gaji karyawan**:
- Jika menggunakan **Mean**, gaji eksekutif yang sangat tinggi akan membuat "rata-rata" terlihat besar â†’ perusahaan bisa mengklaim karyawan dibayar tinggi.  
- Jika menggunakan **Median**, nilai tengah akan mengurangi pengaruh gaji eksekutif, sehingga lebih merefleksikan kondisi mayoritas karyawan.  
- **Kesimpulan:** Pilihan ukuran pusat (Mean atau Median) bisa dipakai untuk mendukung klaim tertentu.  

### Kenapa kadang tetap pilih Mean?
Jika kita mengambil banyak sampel dari populasi, **Mean cenderung lebih konsisten** dibanding ukuran pusat lainnya.  
â¡ï¸ Artinya, Mean lebih stabil untuk analisis statistik inferensial.

---

## ğŸ“Š Distribution Shapes (Bentuk Distribusi)
Distribusi data bisa memiliki bentuk berbeda.  
Tiga bentuk distribusi utama adalah:

### 1. Data Simetris
- Mean â‰ˆ Median â‰ˆ Mode  
- Distribusi berbentuk **lonceng** (normal).  

### 2. Skewed Right (Positively Skewed)
- **Mean > Median**  
- Ekor distribusi panjang di sebelah kanan.  
- Contoh: distribusi harga rumah (beberapa rumah super mahal).  

### 3. Skewed Left (Negatively Skewed)
- **Mean < Median**  
- Ekor distribusi panjang di sebelah kiri.  
- Contoh: distribusi nilai ujian jika banyak siswa mendapat nilai tinggi, tapi ada sedikit yang sangat rendah.  

## ğŸ”‘ Ringkasan
| Bentuk Distribusi   | Hubungan Mean, Median, Mode | Ekor Distribusi | Contoh Data                 |
|---------------------|------------------------------|-----------------|-----------------------------|
| **Symmetric**       | Mean â‰ˆ Median â‰ˆ Mode        | Seimbang        | Tinggi badan                |
| **Positively Skewed** | Mean > Median > Mode        | Ke kanan        | Pendapatan                  |
| **Negatively Skewed** | Mean < Median < Mode        | Ke kiri         | Usia pensiun     

---

## ğŸ“ˆ Hubungan Bentuk Distribusi dengan Ukuran Pusat
- **Mean**: digunakan pada data **interval/rasio** dan **simetris tanpa outlier**.  
- **Median**: digunakan pada data **ordinal, interval, rasio**, dan tetap representatif meski ada outlier.  
- **Mode**: digunakan pada data **nominal, ordinal, interval, rasio**, dan bisa dipakai untuk melihat nilai paling sering muncul.  

---

## ğŸ§® Contoh Python: Mengukur Outlier & Skewness
Kita bisa melihat efek outlier terhadap Mean dan Median.

````python
import numpy as np
import pandas as pd
import statistics

# Data sewa rumah (dalam USD)
data = [850, 900, 950, 1000, 1050, 1100, 1150, 2550]  # ada outlier: 2550

# Hitung ukuran pusat
mean_value = np.mean(data)
median_value = np.median(data)
mode_value = statistics.mode(data)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)

# Gunakan DataFrame untuk analisis tambahan
df = pd.DataFrame({"Harga Sewa": data})
print("\nRingkasan Statistik:")
print(df["Harga Sewa"].describe())

# Skewness
print("\nSkewness:", df["Harga Sewa"].skew())
````

# ğŸ“Š Measures of Spread (Measures of Distribution)

## 1. Apa itu Measures of Spread?
**Measures of Spread** atau **Measures of Distribution** adalah ukuran statistik yang digunakan untuk mengetahui seberapa jauh data tersebar dari pusatnya (mean, median, atau mode).  
Jika **measures of center** hanya memberi tahu "pusat" data, maka **measures of spread** memberi tahu seberapa "variatif" data tersebut.

---

## 2. Pentingnya Measures of Spread
- Membantu memahami **variabilitas** data.
- Berguna untuk membandingkan dua set data dengan rata-rata yang sama, tetapi tingkat penyebaran berbeda.
- Membantu mendeteksi **konsistensi** dan **outlier** dalam data.

---

## 3. Jenis Measures of Spread

### 3.1 Range
- **Definisi**: Selisih antara nilai maksimum dan minimum dalam data.  
- **Rumus**:  
  `Range = Nilai Maksimum - Nilai Minimum`  
- **Kelebihan**: Mudah dihitung.  
- **Kekurangan**: Sangat dipengaruhi oleh outlier.
- Range hanya untuk analisis kasar:
- Kalau Range jauh lebih besar dari IQR atau SD â†’ berarti ada outlier.
- Normal kalau Range â‰ˆ 4â€“6 Ã— SD (kira-kira, untuk data normal).

---

### 3.2 Interquartile Range (IQR)
- **Definisi**: Selisih antara kuartil ketiga (Q3) dan kuartil pertama (Q1).  
- **Rumus**:  
  `IQR = Q3 - Q1`  
- **Kelebihan**: Tidak terpengaruh oleh outlier.  
- **Kegunaan**: Digunakan dalam boxplot untuk melihat sebaran data tengah.
- Untuk data normal, IQR â‰ˆ 1.35 Ã— SD (kira-kira).
- Kalau IQR jauh lebih kecil dari SD â†’ data kemungkinan ada outlier.
- Normal kalau Q1â€“Q3 (IQR) mencakup Â±0.67 SD dari mean.

---

### 3.3 Variance (Varian)
- **Definisi**: Rata-rata kuadrat selisih setiap data terhadap mean.  
- **Rumus populasi**:  
  `Varian (ÏƒÂ²) = ( Î£ (xi - Î¼)Â² ) / N`  

- **Rumus sampel**:  
  `Varian (sÂ²) = ( Î£ (xi - xÌ„)Â² ) / (n - 1)`  

- **Kelebihan**: Mengukur penyebaran data secara matematis.  
- **Kekurangan**: Satuan hasil kuadrat, sehingga sulit ditafsirkan secara langsung.
- Normalitas tetap dicek lewat SD, variance lebih sering dipakai di analisis matematis (misalnya ANOVA, regresi).

- **Keterangan variabel**:

    - ÏƒÂ² = varians populasi (ukuran seberapa jauh data menyebar dari rata-rata).

    - xáµ¢ = nilai data ke-i.

    - Î¼ = mean (rata-rata populasi).

    - N = jumlah data.

    - (xáµ¢ â€“ Î¼) = selisih tiap data dengan rata-rata.

    - (xáµ¢ â€“ Î¼)Â² = selisih itu dikuadratkan (supaya tidak ada bilangan negatif).

    - Î£ = penjumlahan semua selisih kuadrat.
---

### 3.4 Standard Deviation (Simpangan Baku)
- **Definisi**: Akar kuadrat dari variance.  
- **Rumus**:  
  `Standar Deviasi (Ïƒ) = âˆšVarian`  
  `s = âˆšsÂ²`  

- **Kelebihan**: Mudah diinterpretasikan karena satuannya sama dengan data.  
- **Kegunaan**: Paling umum digunakan untuk menggambarkan variasi data.
- Normal distribution â†’ pakai aturan Empiris (68â€“95â€“99.7 Rule):

    - Â±1 SD dari mean â†’ mencakup ~68% data
    - Â±2 SD â†’ mencakup ~95% data
    - Â±3 SD â†’ mencakup ~99.7% data
- Jadi: Kalau banyak data di luar Â±2 SD â†’ kemungkinan ada outlier atau distribusi tidak normal.
- SD relatif "normal" kalau proporsi data sesuai aturan empiris di atas.

---

### 3.5 Mean Absolute Deviation (MAD)
- **Definisi**: Rata-rata dari nilai absolut selisih data terhadap mean.  
- **Rumus**:  
  `MAD = ( Î£ |xi - xÌ„| ) / n`  

- **Kelebihan**: Lebih sederhana dibanding variance.  
- **Kekurangan**: Jarang digunakan dalam statistik formal, tapi bermanfaat untuk analisis cepat.
- Untuk distribusi normal: MAD â‰ˆ 0.8 Ã— SD
- Kalau MAD jauh lebih kecil dari SD â†’ data ada ekor panjang (long tail) atau outlier.

---

## 4. Contoh Kasus
Misalkan data nilai ujian:  70, 75, 80, 85, 90


- **Range** = 90 - 70 = **20**  
- **Median (Q2)** = 80  
- **Q1** = 75, **Q3** = 85 â†’ **IQR** = 85 - 75 = **10**  
- **Mean** = 80  
- **Variance** = 50  
- **Standard Deviation** â‰ˆ 7.07  

ğŸ‘‰ Dari hasil ini terlihat bahwa meskipun mean = 80, penyebaran data tidak terlalu besar (SD hanya 7.07).

---

## 5. Visualisasi
- **Range** â†’ melihat panjang garis data.  
- **IQR** â†’ kotak tengah pada boxplot.  
- **Standard Deviation** â†’ menyebarnya data dari mean pada histogram/kurva distribusi.

## 6. Kesimpulan
- **Range**: ukuran sebaran paling sederhana.  
- **IQR**: tahan terhadap outlier.  
- **Variance & Standard Deviation**: paling sering digunakan, memberikan informasi mendalam.  
- **MAD**: alternatif sederhana dari SD.

# Soal Analisis Data - Measure of Spread

Diketahui data nilai ujian siswa sebagai berikut: 65, 70, 72, 75, 80, 82, 85, 90, 95, 100


### Pertanyaan:

1. Tentukan **range** dari data nilai tersebut.  
2. Hitung **rata-rata (mean)** dan **median** nilai ujian.
3. Hitung **IQR**
4. Hitung  **simpangan baku populasi (Ïƒ)**.  
5. Hitung  **simpangan baku sampel (s)**.
6. Lakukan juga perhitungan empirical rule nya
7. Lakukan analisis dan perhitungan bila perlu dari setiap hasil yang anda dapatkan nilai ujian siswa.
8. Buat plot histogram dan box plot nya

---


````python
import pandas as pd
import numpy as np

# Data nilai ujian siswa
data = [65, 70, 72, 75, 80, 82, 85, 90, 95, 100]
df = pd.DataFrame(data, columns=["Nilai"])

# Hitung statistik dasar
mean = df["Nilai"].mean()
std = df["Nilai"].std(ddof=0)   # SD populasi
std_s = df['Nilai'].std(ddof=1) #SD sample
nilai_range = df["Nilai"].max() - df["Nilai"].min()

# ======================
# Analisis berdasarkan SD (Empirical Rule)
# ======================
intervals = {}
for k in [1, 2, 3]:
    lower = mean - k * std
    upper = mean + k * std
    count = df[(df["Nilai"] >= lower) & (df["Nilai"] <= upper)].shape[0]
    percentage = count / len(df) * 100
    intervals[f"Â±{k} SD"] = {
        "interval": (round(lower,2), round(upper,2)),
        "count": count,
        "percentage": round(percentage,2)
    }

# ======================
# Analisis berdasarkan IQR (Boxplot Rule)
# ======================
Q1 = df["Nilai"].quantile(0.25)
median = df['Nilai'].median()
Q3 = df["Nilai"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Outliers berdasarkan IQR
outliers = df[(df["Nilai"] < lower_bound) | (df["Nilai"] > upper_bound)]

# ======================
# Print hasil
# ======================
print("=== Statistik Dasar ===")
print("Mean:", mean)
print("Standard Deviation:", std)
print("Range:", nilai_range, f"â‰ˆ {round(nilai_range/std,2)} Ã— SD")
print("Q1:", Q1, "Q3:", Q3, "IQR:", IQR)
print("Batas bawah (IQR):", lower_bound, "Batas atas (IQR):", upper_bound)
print("\n=== Distribusi Data (Empirical Rule) ===")
for k, v in intervals.items():
    print(f"{k}: interval {v['interval']} -> {v['count']} data ({v['percentage']}%)")
print("\n=== Outlier Berdasarkan IQR ===")
print(outliers if not outliers.empty else "Tidak ada outlier")
````

````python
# ====== Visualisasi ======

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(12, 5))

# Boxplot
plt.subplot(1, 2, 1)
sns.boxplot(y=df["Nilai"], color="skyblue")
plt.axhline(mean, color="red", linestyle="--", label=f"Mean = {mean:.2f}")
plt.axhline(median, color="green", linestyle="-.", label=f"Median = {median:.2f}")
plt.title("Boxplot Nilai Ujian")
plt.legend()

# Histogram
plt.subplot(1, 2, 2)
sns.histplot(df["Nilai"], bins=5, kde=True, color="lightblue", edgecolor="black")
plt.axvline(mean, color="red", linestyle="--", label=f"Mean = {mean:.2f}")
plt.axvline(median, color="green", linestyle="-.", label=f"Median = {median:.2f}")
plt.title("Histogram Nilai Ujian")
plt.xlabel("Nilai")
plt.ylabel("Frekuensi")
plt.legend()

plt.tight_layout()
plt.show()
````

# Measure of Placement (Ukuran Letak)

## 1. Pendahuluan

**Tujuan:**  
Memahami konsep ukuran letak dalam statistika.

**Mengapa penting:**  
- Membantu memahami **posisi data** dalam distribusi.  
- Berguna untuk **analisis deskriptif**, segmentasi, hingga **data science** (misalnya analisis kuartil, persentil untuk deteksi outlier).  

---

## 2. Konsep Dasar

**Definisi:**  
Measure of Placement (Ukuran Letak) adalah ukuran yang menunjukkan **posisi relatif** suatu data atau sekelompok data dalam distribusi.  

**Perbedaan dengan ukuran lain dalam statistika:**
- **Measure of Central Tendency** (Mean, Median, Mode) â†’ menggambarkan **pusat data**.  
- **Measure of Dispersion** (Range, Variance, Standard Deviation) â†’ menggambarkan **penyebaran data**.  
- **Measure of Placement** (Quartile, Decile, Percentile) â†’ menggambarkan **letak atau posisi relatif data** dalam distribusi.  


# Measure of Placement

## 3. Jenis-Jenis Measure of Placement

### 3.1 Quartiles
- **Definisi:** Membagi data menjadi 4 bagian sama besar.  
- **Simbol:**  
  - Q1 = Kuartil bawah (25%)  
  - Q2 = Median (50%)  
  - Q3 = Kuartil atas (75%)  
- **Rumus posisi (untuk data urut):**  
  Qk = (k * (n + 1)) / 4  
- **Aplikasi:**  
  - Mencari **Interquartile Range (IQR)** = Q3 - Q1  
  - Deteksi **outlier** pada data  

### 3.2 Deciles
- **Definisi:** Membagi data menjadi 10 bagian sama besar.  
- **Simbol:** D1, D2, ..., D9  
- **Rumus posisi (untuk data urut):**  
  Dk = (k * (n + 1)) / 10  
- **Aplikasi:**  
  - Analisis nilai ujian  
  - Distribusi skor siswa  
  - Distribusi pendapatan (income distribution)  

### 3.3 Percentiles
- **Definisi:** Membagi data menjadi 100 bagian sama besar.  
- **Simbol:** P1, P2, ..., P99  
- **Rumus posisi (untuk data urut):**  
  Pk = (k * (n + 1)) / 100  
- **Aplikasi:**  
  - **Pendidikan:** menentukan peringkat siswa  
  - **Data science:** threshold (contoh: P95 untuk deteksi anomali)  
  - **Kesehatan:** growth chart anak  

---

## 4. Metode Perhitungan

### 4.1 Data Tidak Berkelompok (Raw Data)
1. Urutkan data dari terkecil ke terbesar.  
2. Gunakan rumus posisi:  
   - Kuartil ke-k: `Qk = (k * (n + 1)) / 4`  
   - Desil ke-k: `Dk = (k * (n + 1)) / 10`  
   - Persentil ke-k: `Pk = (k * (n + 1)) / 100`  
   (n = banyaknya data)  
3. Jika posisi bukan bilangan bulat â†’ lakukan interpolasi.  

### 4.2 Data Berkelompok (Grouped Data)
Gunakan tabel distribusi frekuensi. Rumus interpolasi linear:  

- Kuartil ke-k:  
  `Qk = L + [ ((k/4 * n) - F) / f ] * c`  

- Desil ke-k:  
  `Dk = L + [ ((k/10 * n) - F) / f ] * c`  

- Persentil ke-k:  
  `Pk = L + [ ((k/100 * n) - F) / f ] * c`  

Keterangan:  
- L = batas bawah kelas kuartil/desil/persentil  
- n = jumlah total data  
- F = frekuensi kumulatif sebelum kelas kuartil/desil/persentil  
- f = frekuensi kelas kuartil/desil/persentil  
- c = panjang kelas

# Measure of Placement dalam Data Science & Data Analysis

## 6. Hubungan dengan Data Science & Data Analysis

Ukuran letak (Quartile, Decile, Percentile) sangat penting dalam analisis data karena:

1. **Deteksi Outlier**  
   - Dengan **IQR (Q3 - Q1)** â†’ nilai di luar `Q1 - 1.5*IQR` atau `Q3 + 1.5*IQR` dianggap outlier.  
   - Sering digunakan pada **preprocessing data** sebelum pemodelan machine learning.  

2. **Feature Engineering**  
   - Percentile dipakai untuk membuat **binning** atau **bucketing** (contoh: segmentasi income).  

3. **Evaluasi Model**  
   - Percentile (misalnya P95, P99) dipakai untuk melihat **skor ekstrim** atau threshold dalam anomaly detection.  

4. **Statistik Deskriptif**  
   - Membantu memahami distribusi data secara cepat, misalnya median (Q2), Q1, dan Q3 untuk membuat boxplot.

## 7. Contoh dengan Python

### Contoh 1: Menghitung Quartile dan IQR
```python
import numpy as np

# Data nilai ujian
data = [55, 60, 62, 65, 70, 72, 75, 80, 82, 90]

# Hitung kuartil
Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)  # Median
Q3 = np.percentile(data, 75)

# Hitung IQR
IQR = Q3 - Q1

print("Q1:", Q1)
print("Q2 (Median):", Q2)
print("Q3:", Q3)
print("IQR:", IQR)

# Deteksi outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Batas bawah:", lower_bound)
print("Batas atas:", upper_bound)
print("Outliers:", outliers)


````python
import numpy as np

# Data nilai ujian
data = [55, 60, 62, 65, 70, 72, 75, 80, 82, 90]

# Hitung kuartil
Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)  # Median
Q3 = np.percentile(data, 75)

# Hitung IQR
IQR = Q3 - Q1

print("Q1:", Q1)
print("Q2 (Median):", Q2)
print("Q3:", Q3)
print("IQR:", IQR)

# Deteksi outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Batas bawah:", lower_bound)
print("Batas atas:", upper_bound)
print("Outliers:", outliers)
````

# Anomaly Detection vs Outlier

## 1. Apa itu Anomaly Detection?
**Anomaly Detection** adalah proses menemukan **pola atau data yang tidak normal** dibandingkan mayoritas data.  
Digunakan dalam **Data Science & Machine Learning** untuk mengidentifikasi data yang menyimpang.  

**Contoh aplikasi:**
- **Fraud detection** â†’ transaksi mencurigakan di bank.  
- **Intrusion detection** â†’ aktivitas mencurigakan di jaringan komputer.  
- **Predictive maintenance** â†’ sensor mesin menunjukkan perilaku aneh.  
- **Monitoring sistem** â†’ response time server yang tiba-tiba melonjak.  

---

## 2. Apa itu Anomaly?
**Anomaly** adalah kejadian atau data yang **tidak sesuai pola biasanya**.  
- Bisa berupa nilai ekstrem, **atau pola aneh** yang tidak terdeteksi hanya dengan statistik sederhana.  

**Contoh:**
- Transaksi Rp 1 miliar di akun yang biasanya transaksi Rp 100 ribu â€“ Rp 2 juta â†’ **anomaly**.  
- User login dari negara berbeda secara tiba-tiba â†’ **anomaly** walaupun nilainya bukan ekstrem.  

---

## 3. Apa itu Outlier?
**Outlier** adalah titik data yang jauh berbeda dari mayoritas data.  
- Biasanya ditentukan dengan **metode statistik**, misalnya:  
  - `Q1 - 1.5*IQR` atau `Q3 + 1.5*IQR`  
  - Z-score  

**Contoh:**
- Tinggi badan rata-rata siswa 150â€“170 cm, ada yang 210 cm â†’ **outlier**.  
- Gaji rata-rata Rp 3â€“10 juta, ada direktur Rp 80 juta â†’ **outlier**, tapi **valid** (bukan error).  

---

## 4. Bedanya Outlier dan Anomaly

| Aspek              | Outlier | Anomaly |
|---------------------|---------|---------|
| **Definisi**        | Nilai ekstrem yang jauh dari mayoritas data. | Data/peristiwa yang menyimpang dari pola normal. |
| **Deteksi**         | Metode statistik (IQR, Z-score, boxplot). | Statistik + Machine Learning (clustering, isolation forest, neural nets, dll). |
| **Konteks**         | Bisa valid (contoh: gaji direktur). | Lebih ke kejadian mencurigakan (fraud, error, serangan). |
| **Contoh**          | Umur 100 tahun di dataset pasien. | Transaksi kartu kredit Rp 1 miliar dari akun yang biasanya kecil-kecil. |

---

## 5. Insight
- Semua **anomaly bisa berupa outlier**.  
- Tidak semua **outlier dianggap anomaly** (karena bisa saja data valid).
